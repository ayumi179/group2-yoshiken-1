```python
import numpy as np
import pandas as pd
from datetime import datetime as dt
```

```python
#各気象データ読み込み
rain_data = pd.read_csv("./kochi_train2015/rain.csv", encoding = 'shift-jis', skiprows = 4)
temp_data = pd.read_csv("./kochi_train2015/temp.csv", encoding = 'shift-jis', skiprows = 3)
wind_data = pd.read_csv("./kochi_train2015/wind.csv", encoding = 'shift-jis', skiprows = 4)
humidity_data = pd.read_csv("./kochi_train2015/humidity.csv", encoding = 'shift-jis', skiprows = 3)
```

```python
rain = rain_data.iloc[:, 0:2]
rain.columns = ['date', 'rain']
#guerilla = rain[rain.rain >= 10]
#guerilla
```

```python
temp_data.columns = ['date', 'temp']
temp = temp_data
```

```python
wind = wind_data.iloc[:, 0:2]
wind.columns = ['date', 'wind']
```

```python
humidity_data.columns = ['date', 'humidity']
humidity = humidity_data
```

```python
temp_df = pd.merge(rain, temp, on='date')
temp_df = pd.merge(temp_df, wind, on='date')
temp_df = pd.merge(temp_df, humidity, on='date')
```
